```python
import pandas as pd

```

```python
crime_data_csv = "data/Clean_Crime_Data.csv"
crime_data = pd.read_csv(crime_data_csv)
neighborhood_csv = 'data/community_nbhd_pop.csv'
neighborhood_df = pd.read_csv(neighborhood_csv)
neighborhood_df = neighborhood_df.rename(columns = {'NEIGHBORHOOD':'Neighborhood', 'COMMUNITY':'Community'}) 
neighborhood_df
```

```python
violent_crimes = ['Burglary of Dwelling', 'Aggravated Robbery', 'Domestic Assault', 'Burglary of Business',
                  'Assault - Dangerous Weapon', 'Robbery of Person', 'Criminal Sexual Conduct', 'Assault',
                  'Robbery of Business', 'Assault of Police Officer', 'Arson', 'Murder', 'Disarm a Police Officer']

crime_data
```

```python
offense_type = crime_data.Description.isin(violent_crimes)
crime_data['Offense Type'] = offense_type
crime_data  
#for index, row in crime_data.iterrows():
#    if row['Description'] in violent_crimes:
#        crime_data.loc[index, 'Offense Type'] = 'Violent'
#    elif row['Description'] not in violent_crimes:
#        crime_data.loc[index, 'Offense Type'] = 'Non-Violent'
        
crime_data
```

```python
crime_data['Offense Type'] = crime_data['Offense Type'].replace({True:'Violent', False:'Non-Violent'})
crime_data
```

```python
crime_data['Neighborhood'] = crime_data['Neighborhood'].str.upper()
```

```python
nonviolent_central_crime = len(nonviolent_crime.loc[nonviolent_crime['Community'] == 'CENTRAL'])
nonviolent_calhoun_crime = len(nonviolent_crime.loc[nonviolent_crime['Community'] == 'CALHOUN-ISLE'])
nonviolent_camden_crime = len(nonviolent_crime.loc[nonviolent_crime['Community'] == 'CAMDEN'])
nonviolent_longfellow_crime = len(nonviolent_crime.loc[nonviolent_crime['Community'] == 'LONGFELLOW'])
nonviolent_nearnorth_crime = len(nonviolent_crime.loc[nonviolent_crime['Community'] == 'NEAR NORTH'])
nonviolent_nokomis_crime = len(nonviolent_crime.loc[nonviolent_crime['Community'] == 'NOKOMIS'])
nonviolent_northeast_crime = len(nonviolent_crime.loc[nonviolent_crime['Community'] == 'NORTHEAST'])
nonviolent_phillips_crime = len(nonviolent_crime.loc[nonviolent_crime['Community'] == 'PHILLIPS'])
nonviolent_powderhorn_crime = len(nonviolent_crime.loc[nonviolent_crime['Community'] == 'POWDERHORN'])
nonviolent_southwest_crime = len(nonviolent_crime.loc[nonviolent_crime['Community'] == 'SOUTHWEST'])
nonviolent_university_crime = len(nonviolent_crime.loc[nonviolent_crime['Community'] == 'UNIVERSITY'])

nonviolent_crime_stat = [nonviolent_calhoun_crime, nonviolent_camden_crime, nonviolent_central_crime, 
                         nonviolent_longfellow_crime, nonviolent_nearnorth_crime, nonviolent_nokomis_crime, 
                         nonviolent_northeast_crime, nonviolent_phillips_crime, nonviolent_powderhorn_crime, 
                         nonviolent_southwest_crime, nonviolent_university_crime]
```

```python
community = ['Calhoun-Isle', 'Camden', 'Central', 'Longfellow', 'Near North', 'Nokomis', 'North East', 'Phillips',
             'Powderhorn', 'Southwest', 'University']
```

```python
summary_crime = pd.DataFrame({'Community':community, 'Population':population, 'Total Crime':total_crime,
                              'Violent Crime':violent_crime_stat, 'Non-Violent Crime':nonviolent_crime_stat,
                              'Total Crime Rate':'', 'Violent Crime Rate':'', 'Non-Violent Crime Rate':''})
summary_crime
```

```python
#Total crime statistics are an aggregation of 5 years of crime data.  I want to get a crime rate for every 1,000 people 
#per year.  So normally I would divide the number of crimes by the population and multiply by 1000 to get the rate 
#per 1000 people.  But because I have five years of crime data, I mulitplied by 200.
summary_crime['Total Crime Rate'] = round(summary_crime['Total Crime']/summary_crime['Population']*200,2)
summary_crime['Violent Crime Rate'] = round(summary_crime['Violent Crime']/summary_crime['Population']*200,2)
summary_crime['Non-Violent Crime Rate'] = round(summary_crime['Non-Violent Crime']/summary_crime['Population']*200,2)
summary_crime
```

```python
summary_crime.to_csv("data/summary_crime.csv", index=True, header=True)
```
